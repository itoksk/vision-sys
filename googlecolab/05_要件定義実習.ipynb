## 4. éæ©Ÿèƒ½è¦ä»¶ã®å®šç¾©

## 5. è¦ä»¶å®šç¾©æ›¸ã®ç”Ÿæˆ

```python
def generate_requirements_document():
    """
    è¦ä»¶å®šç¾©æ›¸ã‚’ç”Ÿæˆã™ã‚‹
    """
    # é¸æŠã•ã‚ŒãŸæ©Ÿèƒ½ã‚’å–å¾—
    selected_must_have = [cb.description for cb in must_checkboxes if cb.value]
    selected_nice_to_have = [cb.description for cb in nice_checkboxes if cb.value]
    
    # è¦ä»¶å®šç¾©æ›¸ã®ãƒ†ãƒ³ãƒ—ãƒ¬ãƒ¼ãƒˆ
    doc = f"""
# å‚·æ¤œå‡ºAIã‚¢ãƒ—ãƒª è¦ä»¶å®šç¾©æ›¸

ä½œæˆæ—¥: {datetime.datetime.now().strftime('%Yå¹´%mæœˆ%dæ—¥')}

## 1. ãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆæ¦‚è¦

### 5W1H
- **Whoï¼ˆèª°ãŒï¼‰**: {saved_5w1h['who']}
- **Whatï¼ˆä½•ã‚’ï¼‰**: {saved_5w1h['what']}
- **Whenï¼ˆã„ã¤ï¼‰**: {saved_5w1h['when']}
- **Whereï¼ˆã©ã“ã§ï¼‰**: {saved_5w1h['where']}
- **Whyï¼ˆãªãœï¼‰**: {saved_5w1h['why']}
- **Howï¼ˆã©ã†ã‚„ã£ã¦ï¼‰**: {saved_5w1h['how']}

## 2. æ©Ÿèƒ½è¦ä»¶

### å¿…é ˆæ©Ÿèƒ½ï¼ˆMust Haveï¼‰
"""
    for feature in selected_must_have:
        doc += f"- [ ] {feature}\n"
    
    doc += "\n### ã‚ã£ãŸã‚‰è‰¯ã„æ©Ÿèƒ½ï¼ˆNice to Haveï¼‰\n"
    for feature in selected_nice_to_have:
        doc += f"- [ ] {feature}\n"
    
    doc += f"""

## 3. éæ©Ÿèƒ½è¦ä»¶

- **åˆ¤å®šé€Ÿåº¦**: {speed_slider.value}ç§’ä»¥å†…
- **åˆ¤å®šç²¾åº¦**: {accuracy_slider.value}%ä»¥ä¸Š
- **åŒæ™‚åˆ©ç”¨ãƒ¦ãƒ¼ã‚¶ãƒ¼æ•°**: {users_slider.value}äºº
- **ä½¿ã„ã‚„ã™ã•**: {usability_dropdown.value}

## 4. åˆ¶ç´„æ¡ä»¶

- **é–‹ç™ºæœŸé–“**: {priority_slider.value}æ™‚é–“ï¼ˆæˆæ¥­æ™‚é–“ï¼‰
- **ä½¿ç”¨æŠ€è¡“**: Pythonã€Teachable Machineã€Google Colab
- **å¯¾è±¡éƒ¨å“**: é‡‘å±éƒ¨å“ï¼ˆ10cmä»¥ä¸‹ï¼‰
- **å‹•ä½œç’°å¢ƒ**: Webãƒ–ãƒ©ã‚¦ã‚¶ï¼ˆChromeæ¨å¥¨ï¼‰

## 5. ãƒ‡ãƒ¼ã‚¿è¦ä»¶

- **è‰¯å“ç”»åƒ**: 100æšä»¥ä¸Š
- **ä¸è‰¯å“ç”»åƒ**: 100æšä»¥ä¸Šï¼ˆæ§˜ã€…ãªå‚·ã®ãƒ‘ã‚¿ãƒ¼ãƒ³ï¼‰
- **ãƒ†ã‚¹ãƒˆç”¨ç”»åƒ**: å„20æšä»¥ä¸Š
- **ç”»åƒå½¢å¼**: JPEGã€PNG
- **ç”»åƒã‚µã‚¤ã‚º**: 640Ã—480ãƒ”ã‚¯ã‚»ãƒ«ä»¥ä¸Šæ¨å¥¨

## 6. æˆåŠŸåŸºæº–

1. æŒ‡å®šã•ã‚ŒãŸç²¾åº¦ï¼ˆ{accuracy_slider.value}%ä»¥ä¸Šï¼‰ã‚’é”æˆã™ã‚‹
2. åˆ¤å®šæ™‚é–“ãŒ{speed_slider.value}ç§’ä»¥å†…ã§ã‚ã‚‹
3. æ¤œæŸ»å“¡ãŒèª¬æ˜ãªã—ã§ä½¿ç”¨ã§ãã‚‹
4. èª¤æ¤œå‡ºç‡ãŒ5%ä»¥ä¸‹ã§ã‚ã‚‹

## 7. ãƒªã‚¹ã‚¯ã¨å¯¾ç­–

| ãƒªã‚¹ã‚¯ | å½±éŸ¿åº¦ | å¯¾ç­– |
|--------|--------|------|
| ãƒ‡ãƒ¼ã‚¿ä¸è¶³ | é«˜ | æ—©æœŸã«ãƒ‡ãƒ¼ã‚¿åé›†ã‚’é–‹å§‹ |
| ç²¾åº¦ãŒç›®æ¨™ã«å±Šã‹ãªã„ | ä¸­ | ãƒ‡ãƒ¼ã‚¿æ‹¡å¼µã€ãƒ¢ãƒ‡ãƒ«èª¿æ•´ |
| é–‹ç™ºæ™‚é–“ã®ä¸è¶³ | ä¸­ | æ©Ÿèƒ½ã®å„ªå…ˆé †ä½ä»˜ã‘ |

## 8. ä»Šå¾Œã®ã‚¹ã‚±ã‚¸ãƒ¥ãƒ¼ãƒ«

1. ãƒ‡ãƒ¼ã‚¿åé›†ï¼ˆ1æ™‚é–“ï¼‰
2. ãƒ¢ãƒ‡ãƒ«ä½œæˆï¼ˆ1æ™‚é–“ï¼‰
3. ã‚¢ãƒ—ãƒªé–‹ç™ºï¼ˆ{priority_slider.value-2}æ™‚é–“ï¼‰
4. ãƒ†ã‚¹ãƒˆãƒ»èª¿æ•´ï¼ˆ30åˆ†ï¼‰
5. ãƒ‰ã‚­ãƒ¥ãƒ¡ãƒ³ãƒˆä½œæˆï¼ˆ30åˆ†ï¼‰
"""
    
    return doc

# è¦ä»¶å®šç¾©æ›¸ç”Ÿæˆãƒœã‚¿ãƒ³
generate_button = widgets.Button(
    description='è¦ä»¶å®šç¾©æ›¸ã‚’ç”Ÿæˆ',
    button_style='primary',
    icon='file-text'
)

output_doc = widgets.Output()

def on_generate_click(b):
    with output_doc:
        output_doc.clear_output()
        doc = generate_requirements_document()
        display(Markdown(doc))
        
        # ãƒ•ã‚¡ã‚¤ãƒ«ã¨ã—ã¦ä¿å­˜
        filename = f"requirements_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.md"
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(doc)
        
        print(f"\nâœ… è¦ä»¶å®šç¾©æ›¸ã‚’ '{filename}' ã¨ã—ã¦ä¿å­˜ã—ã¾ã—ãŸ")
        
        # ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ãƒœã‚¿ãƒ³ã‚’è¡¨ç¤º
        download_button = widgets.Button(
            description='ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰',
            button_style='success',
            icon='download'
        )
        
        def download_file(b):
            files.download(filename)
        
        download_button.on_click(download_file)
        display(download_button)

generate_button.on_click(on_generate_click)
display(generate_button)
display(output_doc)
```

## 7. å„ªå…ˆé †ä½ã®è¨­å®šï¼ˆMoSCoWæ³•ï¼‰

æ©Ÿèƒ½ã®å„ªå…ˆé †ä½ã‚’æ•´ç†ã—ã¾ã™ã€‚

```python
# MoSCoWæ³•ã§ã®å„ªå…ˆé †ä½ä»˜ã‘
print("ğŸ¯ MoSCoWæ³•ã§å„ªå…ˆé †ä½ã‚’è¨­å®š")
print("="*50)

moscow_categories = {
    'Must haveï¼ˆå¿…é ˆï¼‰': [],
    'Should haveï¼ˆã‚ã‚‹ã¹ãï¼‰': [],
    'Could haveï¼ˆã‚ã‚Œã°è‰¯ã„ï¼‰': [],
    'Won\'t haveï¼ˆä»Šå›ã¯å®Ÿè£…ã—ãªã„ï¼‰': []
}

all_features = [
    'ç”»åƒã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰æ©Ÿèƒ½',
    'å‚·ã®æœ‰ç„¡åˆ¤å®š',
    'åˆ¤å®šçµæœã®è¡¨ç¤º',
    'ä¿¡é ¼åº¦ã‚¹ã‚³ã‚¢è¡¨ç¤º',
    'å‚·ã®ä½ç½®è¡¨ç¤º',
    'åˆ¤å®šå±¥æ­´',
    'ãƒãƒƒãƒå‡¦ç†',
    'ãƒ¬ãƒãƒ¼ãƒˆç”Ÿæˆ',
    'ãƒ¦ãƒ¼ã‚¶ãƒ¼ç®¡ç†',
    'APIé€£æº',
    'ãƒ¢ãƒã‚¤ãƒ«å¯¾å¿œ',
    'å¤šè¨€èªå¯¾å¿œ'
]

# ãƒ‰ãƒ©ãƒƒã‚°&ãƒ‰ãƒ­ãƒƒãƒ—é¢¨ã®é¸æŠ
feature_selectors = {}
for feature in all_features:
    selector = widgets.Dropdown(
        options=list(moscow_categories.keys()),
        value='Must haveï¼ˆå¿…é ˆï¼‰' if feature in ['ç”»åƒã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰æ©Ÿèƒ½', 'å‚·ã®æœ‰ç„¡åˆ¤å®š', 'åˆ¤å®šçµæœã®è¡¨ç¤º'] else 'Could haveï¼ˆã‚ã‚Œã°è‰¯ã„ï¼‰',
        description=feature,
        style={'description_width': '200px'}
    )
    feature_selectors[feature] = selector
    display(selector)

# å„ªå…ˆé †ä½ã‚’å¯è¦–åŒ–
visualize_button = widgets.Button(
    description='å„ªå…ˆé †ä½ã‚’å¯è¦–åŒ–',
    button_style='info',
    icon='chart-bar'
)

output_priority = widgets.Output()

def visualize_priority(b):
    with output_priority:
        output_priority.clear_output()
        
        # ã‚«ãƒ†ã‚´ãƒªåˆ¥ã«é›†è¨ˆ
        results = {cat: [] for cat in moscow_categories.keys()}
        for feature, selector in feature_selectors.items():
            results[selector.value].append(feature)
        
        # çµæœã‚’è¡¨ç¤º
        for category, features in results.items():
            print(f"\n### {category}")
            for f in features:
                print(f"  - {f}")
        
        # ã‚°ãƒ©ãƒ•ã§å¯è¦–åŒ–
        fig, ax = plt.subplots(figsize=(10, 6))
        categories = list(results.keys())
        counts = [len(features) for features in results.values()]
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
        
        bars = ax.bar(categories, counts, color=colors)
        ax.set_ylabel('æ©Ÿèƒ½æ•°', fontsize=12)
        ax.set_title('æ©Ÿèƒ½ã®å„ªå…ˆé †ä½åˆ†å¸ƒ', fontsize=14)
        
        # æ•°å€¤ã‚’è¡¨ç¤º
        for bar, count in zip(bars, counts):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
                   str(count), ha='center', va='bottom')
        
        plt.xticks(rotation=15)
        plt.tight_layout()
        plt.show()

visualize_button.on_click(visualize_priority)
display(visualize_button)
display(output_priority)
```
